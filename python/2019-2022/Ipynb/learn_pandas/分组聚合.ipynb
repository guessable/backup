```python
# 自定义聚合函数，应用与每一组
def max_sub_min(df):
    '''
    max - min
    '''
    return df.max()-df.min()

grouped['total_bill'].agg(max_sub_min)
```

## apply

```python
def top(df,n=5,columns='tip'):
    '''
    返回列top n 所在行
    '''
    return df.sort_values(by=columns)[-n:]
```

```python
top(tips)
```

```python
grouped = tips.groupby('smoker')
```

```python
grouped.apply(top) # top会应用在每个分组上，然后concat,传给apply的函数只需返回一个pandas对象或标量值即可
```

```python
tips.groupby(['smoker','sex']).apply(top,n=2,columns='total_bill') 
# 如果传给apply的函数能够接受其他参数或关键字，则可以将这些内容放在函数名后面一并传入
```

```python
tips.groupby('smoker',group_keys=False).apply(top,n=3) #禁止分组
```
